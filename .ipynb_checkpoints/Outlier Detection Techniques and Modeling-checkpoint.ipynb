```python
## Univariate (one variable) Outlier Techniques
```

```python
## Method 1: inter-quartile range
```

```python
plt.figure(figsize = (2,4))
sns.boxplot(y = df['math score'])
```

```python
## figure above shows that there are in fact some scores below the lower whisker, proving there are outliers
```

```python
# fuction to find outliers
def IQR(df, column):
    global lower, upper
    q25, q75= np.quantile(df[column], 0.25), np.quantile(df[column], 0.75)
    #formula for iqr
    iqr = q75 - q25
    #cuttoff for outliers
    cut_off = iqr * 1.5 # length of the whisker
    lower, upper = q25 - cut_off, q75 + cut_off
    # number of outliers above and below the IQR
    df1 = df[df[column] > upper]
    df2 = df[df[column] < lower]
    print('IQR is', iqr)
    print('Lower bound is', lower)
    print('Upper bound is', upper)
    print('Total below is', df2.shape[0])
    print('Total above is', df1.shape[0])
    print('Total numer of outliers are:', df1.shape[0] + df2.shape[0]) # adds up the total and prints them
    return 
```

```python
IQR(df, 'math score')
```

```python
# through this method, we have found the four outliers
```

```python
# visualizing the outliers
plt.figure(figsize = (10,5))
sns.distplot(df['math score'], kde=False)
plt.axvspan(xmin = lower, xmax = df['math score'].min(), alpha=0.1, color = 'red')
plt.axvspan(xmin = upper, xmax = df['math score'].max(), alpha=0.1, color = 'red')
```

```python
# as you can see, the students that scored below 28 on the math exam are clear outliers from the rest of the data
```

```python
# creating a new data frame with the outliers removed
df_new = df[(df['math score'] > lower) & (df['math score'] < upper)]
plt.figure(figsize = (10,5))
sns.distplot(df_new['math score'], kde=False)
```

```python
### Method 2: Standard Deviation
```

```python
plt.figure(figsize=(10,5))
sns.distplot(df['math score'])
```

```python
# left skewed
```

```python
def std(df, column):
    global lower, upper
    mean, std = df[column].mean(), df[column].std()
    # cutoff 
    cut_off = std * 3 # 3 standard deviations from the mean
    # bounds
    lower, upper= mean - cut_off, mean + cut_off   
    df1 = df[df[column] > upper]
    df2 = df[df[column] < lower]
    print('Lower bound is', lower)
    print('Upper bound is', upper)
    print('Total below is', df2.shape[0])
    print('Total above is', df1.shape[0])
    print('Total numer of outliers are:', df1.shape[0] + df2.shape[0]) # adds up the total and prints them
    return 
```

```python
std(df, 'math score')
```

```python
# visualizing the outliers
plt.figure(figsize = (10,5))
sns.distplot(df['math score'], kde=False)
plt.axvspan(xmin = lower, xmax = df['math score'].min(), alpha=0.1, color = 'red')
plt.axvspan(xmin = upper, xmax = df['math score'].max(), alpha=0.1, color = 'red')
```

```python
# to better grasp this, let's look at the lowest math scores 
math = df['math score']
math.sort_values(ascending=True)
```

```python
## as we can see there are 2 outliers below the lower bound
```

```python
## so far, the results for IQR and STD have been different. Why?
## one reason may be because STD is heavily reliant on the mean and STD, meaning that extreme outliers might increase
## the STD to a point where it fails to detect moderate outliers.
```

```python
## Does this mean we cannot use STD to reliably detect outliers at all? Not necessarily.
## Ideally, the STD test is used on Gaussian distribution (bell shaped). This means that in an ideal scenario, 1 STD should
## account for approximately 68% of the data. In our case, while it is a Gaussian distribution, one STD only covers 66.5%,
# which is why it may be slightly off compared to the IQR method.
```

```python
## Method 3: Z-Score
## similar to the STD method, this method assumes a Gaussian distribution of data, which is what we are working with
```

```python
def zscore(data):
    global outliers, zscore
    outliers =[]
    zscore = []
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score = (i-mean)/std
        zscore.append(z_score)
        if (np.abs(z_score) > threshold):
            outliers.append(i)
    print('Total number of outliers is', len(outliers))
    return    
```

```python
zscore(df['math score'])
```

```python
plt.figure(figsize = (10,5))
sns.distplot(zscore)
plt.axvspan(xmin = 3 ,xmax= max(zscore),alpha=0.2, color='red')
```

```python
# as expected, similar to the STD method, the z score method also only detects 2 outliers because both are heavily dependent
# on mean and standard deviation in their calculations
```

```python
## While univariate outlier tests on this dataset helped to remove some exam scores that may not be beneficial for analysis,
## we have yet to utilize these tests to interpret the data in meaningful ways. To better understand the data and understand
## why students may have scored the way they did, it is beneficial to do a multi-dimensional analysis.
```

```python
## Method 1: DBSCAN
```

```python
## Is there a correlation between reading and writing scores?
```

```python
# DBSCAN clusters values together based on how close they are to each other
x = df[['reading score', 'writing score']].values # all the values for reading and writing scores
db = DBSCAN(eps=5, min_samples=10).fit(x) #eps clusters values within a range from each other
# value of min_samples is however many different data points are required to form a cluster
# data points that do not belong to any cluster are considered outliers in the dataset
labels=db.labels_
```

```python
pd.Series(labels).value_counts() # -1 depicts the number of outliers
```

```python
plt.figure(figsize=(30,30))
lab = set(labels)
colors = ['green', 'red']
for color, label in zip(colors, lab):
    mask = [True if l == label else False for l in labels]
    plt.plot(x[:,0][mask], x[:, 1][mask], 'o', color=color)
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
```

```python
## as you can see based on this graph, there is a direct correlation between reading and writing scores
## the number of outliers will change based on the values of eps and min_samples
## in my case, I defined a value to be in a cluster if it was within a score of 5 from another score, and 10 values needed to 
## be grouped together to be considered a cluster. 
## the downside to this method is that it can be subjective, as we have to make assumptions such as estimating the right values
## for eps and min_samples
```
