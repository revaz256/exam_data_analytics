```python
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()) #since x is a series we can use .str
```

```python
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').value_counts(normalize = True)).loc[:, True]
# percentage version of above 
```

```python
num_series = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
pct_series = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').value_counts(normalize = True)).loc[:, True]
respondents = df['Country'].value_counts()
python_df = pd.concat([respondents, num_series, pct_series], axis='columns', sort=False)
```

```python
python_df.rename(columns={'Country': 'TotalRespondents', 'LanguageWorkedWith': 'UsePython' }, inplace=True)
python_df.head(50)
```

```python
python_df.sort_values('TotalRespondents', ascending=False)
```
